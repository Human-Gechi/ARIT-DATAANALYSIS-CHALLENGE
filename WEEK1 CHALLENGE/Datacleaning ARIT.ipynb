```python
import pandas as pd
import re
df = pd.read_csv('data.csv', encoding='utf-8', encoding_errors='ignore')
df
```

```python
df.info()
```

```python

replacing = re.compile(r'[a-zA-Z]+')
df[['InvoiceNo', 'StockCode']] = df[['InvoiceNo', 'StockCode']].apply(lambda col: col.str.replace(replacing, '', regex=True))
df[['InvoiceNo', 'StockCode']] = df[['InvoiceNo', 'StockCode']].apply(lambda col: col.str.strip())
df[["InvoiceNo", "StockCode"]] = df[["InvoiceNo", "StockCode"]].apply(pd.to_numeric, errors='coerce')
df["InvoiceNo"] = df["InvoiceNo"].fillna(0).astype(int)
df["StockCode"] = df["StockCode"].fillna(0).astype(int)
df.iloc[140:150]
```

```python
df["CustomerID"] = pd.to_numeric(df["CustomerID"], errors='coerce')
df["CustomerID"] = df["CustomerID"].fillna(0).astype(int)
df
```

```python
df['Quantity'] = df["Quantity"].abs()
df['UnitPrice'] = df["UnitPrice"].abs()
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df['TotalPrice'] = df['TotalPrice'].abs()
df['TotalPrice'] = df['TotalPrice'].fillna(0)
df['TotalPrice'] = df['TotalPrice'].astype(float)
df["TotalPrice"] = df["TotalPrice"].round(2)
```

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
```

```python
df.to_csv('cleaned_data.csv', index=False, encoding='utf-8')
```
